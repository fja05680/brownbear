# List of DJIA (DOW 30) companies

Retreive from https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average

output: 'dow30.csv'

```python
from datetime import datetime

import pandas as pd

from brownbear.utility import get_wikipedia_table

pd.options.mode.chained_assignment = None  # default='warn'
pd.set_option('display.max_rows', 600)

# -*- encoding: utf-8 -*-
%matplotlib inline
```

```python
%%javascript
IPython.OutputArea.prototype._should_scroll = function(lines) {
    return false;
}
```

```python
get_table = get_wikipedia_table
```

```python
title = 'Dow Jones Industrial Average'
filename = 'dow30.csv'
dow30 = get_table(title, filename, match='Symbol')

# dd/mm/YY H:M:S
now = datetime.now()
dt_string = now.strftime("%m/%d/%Y %H:%M:%S")
print('{} (retrieved {})'.format(title, dt_string))
dow30
```
