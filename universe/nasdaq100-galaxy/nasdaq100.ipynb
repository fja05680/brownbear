# List of Nasdaq 100 companies

Retreive from https://en.wikipedia.org/wiki/NASDAQ-100#Components

output: 'nasdaq100.csv'

```python
# Imports
from datetime import datetime
import numpy as np
import pandas as pd
import os
import re
import shutil
import wikipedia as wp

pd.options.mode.chained_assignment = None  # default='warn'
pd.set_option('display.max_rows', 200)

# -*- encoding: utf-8 -*-
%matplotlib inline
```

```python
%%javascript
IPython.OutputArea.prototype._should_scroll = function(lines) {
    return false;
}
```

```python
def get_table(title, filename, match, use_cache=False):

    if use_cache and os.path.isfile(filename):
        pass
    else:
        html = wp.page(title).html()
        df = pd.read_html(html, header=0, match=match)[0]
        
        df.to_csv(filename, header=True, index=False, encoding='utf-8')
            
    df = pd.read_csv(filename)
    return df
```

```python
title = 'NASDAQ-100'
filename = 'nasdaq100.csv'
sp400 = get_table(title, filename, match='Ticker')

# dd/mm/YY H:M:S
now = datetime.now()
dt_string = now.strftime("%m/%d/%Y %H:%M:%S")
print('{} (retrieved {})'.format(title, dt_string))
sp400
```
