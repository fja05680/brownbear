# List of S&P 400 companies

Retreive from https://en.wikipedia.org/wiki/List_of_S%26P_400_companies

output: 'sp400.csv'

```python
from datetime import datetime

import numpy as np
import pandas as pd

from brownbear.utility import get_wikipedia_table

pd.options.mode.chained_assignment = None  # default='warn'
pd.set_option('display.max_rows', 600)

# -*- encoding: utf-8 -*-
%matplotlib inline
```

```python
%%javascript
IPython.OutputArea.prototype._should_scroll = function(lines) {
    return false;
}
```

```python
get_table = get_wikipedia_table
```

```python
title = 'List of S&P 400 companies'
filename = 'sp400.csv'
sp400 = get_table(title, filename, match='Headquarters')

# dd/mm/YY H:M:S
now = datetime.now()
dt_string = now.strftime("%m/%d/%Y %H:%M:%S")
print('{} (retrieved {})'.format(title, dt_string))
sp400
```
