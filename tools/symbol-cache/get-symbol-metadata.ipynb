# Get Symbol Metadata

Get metadata of any or all symbols in symbol_cache
    

```python
import pandas as pd
import brownbear as bb

# format price data
pd.options.display.float_format = '{:0.2f}'.format

# increase display of dataframe rows
pd.set_option('display.max_rows', 2000)
```

Get metadata for the symbols below

```python
df = bb.get_symbol_metadata(symbols=['msft', 'orcl', 'tsla'])
df.sort_values('num_years', ascending=False).reset_index(drop=True)
```

Get metadata for all symbols in the cache directory

```python
df = bb.get_symbol_metadata()
df.sort_values('num_years', ascending=False).reset_index(drop=True)
```
